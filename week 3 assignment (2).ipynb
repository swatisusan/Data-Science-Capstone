```python
#Question 2

#import libraries
import requests
import pandas as pd
import csv
import urllib.request
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
response = requests.get(url)
print(response.status_code)
soup = BeautifulSoup(response.text,"html.parser")
table = soup.findAll('table',{"class":"sortable"})[0]
tr = table.findAll(['tr'])[0:185]
csvFile = open("postalcode.csv",'wt',newline='', encoding='utf-8')
writer = csv.writer(csvFile)  
try:   
        for cell in tr:
            th = cell.find_all('th')
            th_data = [col.text.strip('\n') for col in th]
            td = cell.find_all('td')
            row = [i.text.replace('\n','') for i in td]
            writer.writerow(th_data+row)      
        
finally:   
    csvFile.close()
    
#read the data into a dataframe    
    
df= pd.read_csv("postalcode.csv")
df
```

```python
# Question 3
# cleaning the data - removing the data that has value as not assigned in the Borough column
df1=df[df.Borough!="Not assigned"].reset_index(drop=True)
df1
```

```python
#grougp the data by borough
df_grouped = df1.groupby(["Postal Code", "Borough"], as_index=False).agg(lambda x: ", ".join(x))
df_grouped.head()
```

```python
#replace the not assigned value in Neighbourhood with the corresspong borough value
df_grouped["Neighbourhood"]=df_grouped["Neighbourhood"].replace("Not assigned",df_grouped["Borough"])
```
